```python
import cv2
import os
import argparse
import numpy as np
from tqdm import tqdm
#from PIL import Image

PIC_SIZE = 256

path = '/media/ouc/4T_B/gds/data/ParisView/paris_train_original/'
output = '/media/ouc/4T_B/gds/data/ParisView/image-256-3part/'

imgs = []
result_imgs = []

if (not os.path.isdir(output)):
    os.mkdir(output)

for i,file in tqdm(enumerate(sorted(os.listdir(path)))):
    img = cv2.imread(os.path.join(path, file))
    h = img.shape[0]
    w = img.shape[1]
    img_l = img[:, 0:h]
    img_r = img[:, w-h:]
    img_c = img[:, (w-h)//2:(w+h)//2]

    # print(img_l.shape, img_r.shape, img_c.shape)

    img_l = cv2.resize(img_l, (PIC_SIZE, PIC_SIZE))
    img_r = cv2.resize(img_r, (PIC_SIZE, PIC_SIZE))
    img_c = cv2.resize(img_c, (PIC_SIZE, PIC_SIZE))


    cv2.imwrite(output + "/%05d" % i + "_l.png", img_l)
    cv2.imwrite(output + "/%05d" % i + "_r.png", img_r)
    cv2.imwrite(output + "/%05d" % i + "_c.png", img_c)
    
    
```
